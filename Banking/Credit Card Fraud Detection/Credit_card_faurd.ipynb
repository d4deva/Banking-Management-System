```python
X= df1.drop('Risk',axis=1)
y=df1.Risk
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=55)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight="balanced")
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_true=y_test,y_pred=y_pred)
```

```python
print('Accuracy Score:',accuracy_score(y_test,y_pred)*100,'%')
```

```python
print(classification_report(y_test,y_pred))
```

## Knn Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
score = {}
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    score[i] = knn.score(X_test, y_test)
print (score)
```

```python
error = {}
# Calculating error for K values between 1 and 40
for i in range(1, 40):  
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error[i]=np.mean(pred_i != y_test)
print (error)
```

```python
x = list(score.keys())
y=list(score.values())
plt.figure(figsize = (11, 4))
plt.subplot(1,2,1,)
plt.plot(x, y, color='blue', linestyle='dashed', marker='o',  
         markerfacecolor='red', markersize=8)
plt.xlabel('k')
plt.ylabel('Accuracy')

plt.subplot(1,2,2)
plt.plot(list(error.keys()), list(error.values()), color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=8)
plt.xlabel('k')
plt.ylabel('Error Rate')

plt.show()
```

```python
print(max(score, key=score.get), max(score.values()))
print(min(error, key=error.get), min(error.values()))
```

## Random Forest Classifier

```python
X = df1.drop('Risk',axis=1)
y = df1.Risk
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
y_pred2 = rfc.predict(X_test)
```

```python
confusion_matrix(y_true=y_test,y_pred=y_pred2)
```

```python
print('Accuracy Score:',accuracy_score(y_test,y_pred2)*100,'%')
```

```python
rfc.feature_importances_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 110, num = 11)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
#Create the random grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Use the random grid to search for best hyperparameters

# First create the base model to tune
rf = RandomForestClassifier()

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = param_grid, 
                               n_iter = 100, cv = 5, verbose=2, random_state = 10, n_jobs = -1)

# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
# Manually provide the best parameters to model for training
rf_class= RandomForestClassifier(**{'n_estimators': 600,
 'min_samples_split': 5,
 'min_samples_leaf': 1,
 'max_features': 'auto',
 'max_depth': 44,
 'bootstrap': False})
```

```python
rf_class.fit(X_train, y_train)
```

```python
y_pred3 = rf_class.predict(X_test)
```

```python
confusion_matrix(y_true=y_test,y_pred=y_pred3)
```

```python
print('Accuracy Score:',accuracy_score(y_test,y_pred3)*100,'%')
```

```python
print(classification_report(y_test,y_pred3))
```

```python
# pickling the Model
import pickle
file = open('credit_card_fraud.pkl', 'wb')
pickle.dump(rf_class, file)
```
